```python
import matplotlib.pyplot as plt

data = {
    2010: 19,
    2011: 15,
    2012: 10,
    2013: 17,
    2014: 9,
    2015: 10,
    2016: 18,
    2017: 18,
    2018: 17,
    2019: 16,
    2020: 15,
    2021: 15,
    2022: 13,
    2023: 12,
    2024: 11,
    2025: 5
}

years = list(data.keys())
counts = list(data.values())

plt.figure(figsize=(10, 5), dpi=300)
plt.plot(years, counts, marker='o', color='#13452D', linewidth=2, markerfacecolor='#13452D')

plt.xlabel("Year", fontweight='bold', color='#13452D')
plt.ylabel("Video's duration", fontweight='bold', color='#13452D')
plt.xticks(years, rotation=45, fontweight='bold', color='#13452D')
plt.yticks(range(0, max(counts) + 2), fontweight='bold', color='#13452D')

ax = plt.gca()
for spine in ax.spines.values():
    spine.set_color('#13452D')
    spine.set_linewidth(1.5)

plt.grid(False)

plt.show()
```

```python
import matplotlib.pyplot as plt

# Données : nombre de vidéos par saison
data = {
    'Printemps': 120,
    'Été': 180,
    'Automne': 90,
    'Hiver': 110
}

labels = list(data.keys())
sizes = list(data.values())
colors = ['#A8E6CF', '#FFD3B6', '#FFAAA5', '#DCE775']  # Couleurs douces par saison
explode = (0.05, 0.05, 0.05, 0.05)  # Séparation légère pour chaque section

plt.figure(figsize=(6, 6), dpi=300)
plt.pie(
    sizes,
    labels=labels,
    autopct='%1.1f%%',
    startangle=90,
    colors=colors,
    explode=explode,
    textprops={'fontweight': 'bold', 'color': '#13452D'}
)

plt.title("Saisonnalité des publications des vidéos", fontweight='bold', color='#13452D')
plt.axis('equal')

plt.show()
```

```python
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature

# Coordonnées de 15 localisations en France (latitude, longitude)
locations = {
    'Paris': (48.8566, 2.3522),
    'Lyon': (45.7640, 4.8357),
    'Marseille': (43.2965, 5.3698),
    'Lille': (50.6292, 3.0573),
    'Toulouse': (43.6047, 1.4442),
    'Bordeaux': (44.8378, -0.5792),
    'Nantes': (47.2184, -1.5536),
    'Nice': (43.7102, 7.2620),
    'Strasbourg': (48.5734, 7.7521),
    'Rennes': (48.1173, -1.6778),
    'Grenoble': (45.1885, 5.7245),
    'Dijon': (47.3220, 5.0415),
    'Le Havre': (49.4944, 0.1079),
    'Reims': (49.2583, 4.0317),
    'Clermont-Ferrand': (45.7772, 3.0870)
}

# Valeurs associées à chaque localisation
values = [23, 15, 34, 12, 40, 18, 22, 25, 30, 17, 20, 16, 19, 21, 28]

# Création de la carte
plt.figure(figsize=(10, 10))
ax = plt.axes(projection=ccrs.PlateCarree())
ax.set_extent([-5, 10, 41, 52], crs=ccrs.PlateCarree())

# Ajouter des éléments géographiques
ax.add_feature(cfeature.BORDERS, linewidth=1)
ax.add_feature(cfeature.COASTLINE, linewidth=0.5)
ax.add_feature(cfeature.LAND, facecolor='#f0f0f0')
ax.add_feature(cfeature.OCEAN, facecolor='#a0c8f0')

# Afficher les bulles avec les valeurs
for (city, (lat, lon)), val in zip(locations.items(), values):
    ax.scatter(lon, lat, s=val * 10, color='#13452D', alpha=0.7, transform=ccrs.PlateCarree())
    ax.text(lon, lat, str(val), fontsize=9, fontweight='bold', ha='center', va='center',
            color='white', transform=ccrs.PlateCarree())

plt.title("Cartographie des localisations en France", fontsize=14, fontweight='bold', color='#13452D')
plt.tight_layout()
plt.show()
```

```python
import folium
from folium.features import DivIcon

locations = [
    ("Paris", 48.8566, 2.3522, 77),
    ("Lyon", 45.7640, 4.8357, 50),
    ("Marseille", 43.2965, 5.3698, 40),
    ("Lille", 50.6292, 3.0573, 120),
    ("Toulouse", 43.6047, 1.4442, 30),
    ("Bordeaux", 44.8378, -0.5792, 80),
    ("Nantes", 47.2184, -1.5536, 37),
    ("Nice", 43.7102, 7.2620, 23),
    ("Strasbourg", 48.5734, 7.7521, 66),
    ("Rennes", 48.1173, -1.6778, 70),
    ("Grenoble", 45.1885, 5.7245, 50),
    ("Dijon", 47.3220, 5.0415, 33),
    ("Le Havre", 49.4944, 0.1079, 35),
    ("Reims", 49.2583, 4.0317, 78),
    ("Clermont-Ferrand", 45.7772, 3.0870, 67)
]

map_france = folium.Map(location=[46.8, 2.5], zoom_start=6, tiles="CartoDB positron")

for city, lat, lon, value in locations:
    folium.CircleMarker(
        location=[lat, lon],
        radius=value / 2,
        color="#13452D",
        fill=True,
        fill_color="#13452D",
        fill_opacity=0.8
    ).add_to(map_france)

    folium.map.Marker(
        [lat, lon],
        icon=DivIcon(
            icon_size=(150,36),
            icon_anchor=(0,0),
        )
    ).add_to(map_france)

map_france.save("carte_france_bulles.html")
```

```python
import matplotlib.pyplot as plt

paillage = {
    2010: 5, 2011: 6, 2012: 7, 2013: 9, 2014: 10, 2015: 12,
    2016: 14, 2017: 15, 2018: 15, 2019: 16, 2020: 17,
    2021: 18, 2022: 18, 2023: 19, 2024: 19, 2025: 20
}

jardin_foret = {
    2010: 1, 2011: 2, 2012: 3, 2013: 4, 2014: 6, 2015: 7,
    2016: 9, 2017: 11, 2018: 13, 2019: 14, 2020: 15,
    2021: 15, 2022: 16, 2023: 17, 2024: 18, 2025: 18
}

years = list(paillage.keys())
paillage_counts = list(paillage.values())
jardin_counts = list(jardin_foret.values())

plt.figure(figsize=(10, 5), dpi=300)

plt.plot(years, paillage_counts, marker='o', color='#13452D',
         linewidth=2, markerfacecolor='#13452D', label="Paillage")

plt.plot(years, jardin_counts, marker='s', color='#5E8B7E',
         linewidth=2, markerfacecolor='#5E8B7E', label="Jardin-forêt")

plt.xlabel("Year", fontweight='bold', color='#13452D')
plt.ylabel("Technique presence", fontweight='bold', color='#13452D')
plt.xticks(years, rotation=45, fontweight='bold', color='#13452D')

max_y = max(max(paillage_counts), max(jardin_counts))
plt.yticks(range(0, max_y + 2), fontweight='bold', color='#13452D')

ax = plt.gca()
for spine in ax.spines.values():
    spine.set_color('#13452D')
    spine.set_linewidth(1.5)

plt.legend(frameon=False, loc="upper left", fontsize=9)
plt.grid(False)

plt.tight_layout()
plt.show()
```

```python
import psycopg
import os
import matplotlib.pyplot as plt
import numpy as np

conn = psycopg.connect(
    dbname="youtubestay",
    user="postgres",
    password=os.getenv("POSTGRE_PASSWORD"),
    host="localhost",
    port="5432"
)

cur = conn.cursor()

cur.execute("""
    SELECT nombre_abonnes_total FROM chaines_metriques;
""")

rows = cur.fetchall()
subscribers = [row[0] for row in rows]

cur.close()
conn.close()

Q1 = np.percentile(subscribers, 25)
Q3 = np.percentile(subscribers, 75)
IQR = Q3 - Q1
whisker_low = Q1 - 1.5 * IQR
whisker_high = Q3 + 1.5 * IQR

print(f"Min : {min(subscribers)}")
print(f"Q1 : {Q1}")
print(f"Médiane : {np.percentile(subscribers, 50)}")
print(f"Q3 : {Q3}")
print(f"Max : {max(subscribers)}")
print(f"IQR : {IQR}")
print(f"Whisker bas (Q1 - 1.5*IQR) : {whisker_low}")
print(f"Whisker haut (Q3 + 1.5*IQR) : {whisker_high}")

plt.boxplot(subscribers, whis=1.5)
plt.title("Distribution du nombre d'abonnés")
plt.ylabel("Nombre d'abonnés")
plt.show()
```
